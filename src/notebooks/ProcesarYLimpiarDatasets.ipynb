### Tratar accidentes con víctimas

```python
xls_series_historicas = pd.ExcelFile('../data/original/Series-Historicas-Accidentes-30-dias-2019.XLSX')
```

<a id="excels_fallecidosXmeses"></a>
### Tratar número de fallecidos por meses

```python
# Leer la hoja EXCEL saltándonos las filas inicales hasta llegar a la fila con las cabeceras que nos interesan
series_historicas_fallecidosXmes = pd.read_excel(xls_series_historicas,'M_I-U_Meses', header=2)
series_historicas_fallecidosXmes
```

```python
# Eliminar las últimas filas de la hoja sin información
series_historicas_fallecidosXmes.dropna(thresh=4, inplace=True)
series_historicas_fallecidosXmes
```

```python
series_historicas_fallecidosXmes.dtypes
```

```python
# Crear un dataset con un índice temporal por mes
fallecidosXmes_df = pd.DataFrame({"Fecha" :[], "Fallecidos":[]})
fallecidosXmes_df
filaAux = 0
for fila in range(len(series_historicas_fallecidosXmes)):
    anio = series_historicas_fallecidosXmes.iat[fila,0]
    for columna in range(1,13):
        fecha = pd.to_datetime(f"{anio}-{columna:02d}")
        fallecidosXmes_df.loc[filaAux] = [fecha, series_historicas_fallecidosXmes.iat[fila,columna]]
        filaAux += 1

fallecidosXmes_df
```

```python
fallecidosXmes_df.to_csv('../data/clean/fallecidos_x_meses_1993_2019.csv', index=False)
```

<a id="excels_accidentesXprovincia"></a>
### Tratar número de accidentes por provincia

```python
# Leer la hoja EXCEL saltándonos las filas inicales hasta llegar a la fila con las cabeceras que nos interesan
series_historicas_accidentesXprov = pd.read_excel(xls_series_historicas,'Acc_Vict_I-U_Provincias', header=2)
series_historicas_accidentesXprov
```

```python
# Eliminar las últimas filas, que no se corresponden con una provincia
series_historicas_accidentesXprov = series_historicas_accidentesXprov.loc[0:51]
series_historicas_accidentesXprov
```

```python
def cambiar_nombre_provincia(nombre):
    nuevo_nombre = nombre
    try:
        ind = nombre.index('/')
    except:
        ind = -1
    if ind > 0:
        if nombre == "Araba/Álava":
            nuevo_nombre = "Álava"
        else:
            nuevo_nombre = nombre[:ind]
    elif nombre == "Santa Cruz de Tenerife":
        nuevo_nombre = "Sta Cruz Tenerife"
        
    return nuevo_nombre

# Renombramos los nombre de provincias largos para que salgan mejor al etiquetar ejes de gráficas
series_historicas_accidentesXprov['Provincias'] = series_historicas_accidentesXprov['Provincias'].apply(cambiar_nombre_provincia)
series_historicas_accidentesXprov
```

```python
series_historicas_accidentesXprov['Total'] = series_historicas_accidentesXprov.apply(lambda x: x[1:].sum(), axis = 1)
```

```python
series_historicas_accidentesXprov['CodProvincia'] = [f'{x:02d}' for x in np.arange(1,53)]
```

```python
series_historicas_accidentesXprov
```

```python
#series_historicas_accidentesXprov
series_historicas_accidentesXprov.to_csv('../data/clean/historico_accidentes_x_provincia_1993_2019.csv', index=False)
```

## Tratar número de conductores implicados en accidentes con víctimas por sexo 

```python
import glob
import sys
sys.path.append(r"../")
import utils.ProcesadoDatos as procdat
```

```python
# Crear un objeto de la clase ProcesadorDFConductores que implementa metodos para tratar los ficheros EXCELs
# con información de conductores implicados en accidentes con víctimas por sexo
procesadorDFConductores = procdat.ProcesadorDFConductores() #(True)
# Obtener una lista con las rutas de los EXCELs
list_ficheros_excel = glob.glob('../data/original/GRUPO-4*')
list_ficheros_excel.sort(key=procdat.get_anio_excel)
list_ficheros_excel;
```

```python
lista_conductores_x_sexo = []

# Bucle para tratar uno a uno los EXCELs de cada año
for fichero_excel in list_ficheros_excel:
    xls_conductores_x_sexo = pd.ExcelFile(fichero_excel)
    anio = procdat.get_anio_excel(fichero_excel)
    # Obtener el nombre de la hoja con información a nivel de vías interurbanas
    nombrehojaI = procesadorDFConductores.anio2nombrehoja(anio, True)
    # Obtener el nombre de la hoja con información a nivel de vías urbanas
    nombrehojaU = procesadorDFConductores.anio2nombrehoja(anio, False)
    # print(nombrehojaI, nombrehojaU)
    try:
        # Crear un DataFrame con información a nivel de vías interurbanas
        conductores_por_sexo = pd.read_excel(xls_conductores_x_sexo, nombrehojaI)
        conductores_por_sexo_I = procesadorDFConductores.get_conductores_x_sexo_anios(conductores_por_sexo, anio)
        #display(conductores_por_sexo_I)
        # Crear un DataFrame con información a nivel de vías urbanas
        conductores_por_sexo = pd.read_excel(xls_conductores_x_sexo, nombrehojaU)
        conductores_por_sexo_U = procesadorDFConductores.get_conductores_x_sexo_anios(conductores_por_sexo, anio)
        #display(conductores_por_sexo_U)
        # Crear un DataFrame único para un año uniendo la información de vías interurbanas y de vías urbanas
        conductores_por_sexo = procesadorDFConductores.unir_conductores_x_sexo_anio(conductores_por_sexo_I, conductores_por_sexo_U)
        # display(conductores_por_sexo)
        lista_conductores_x_sexo.append(conductores_por_sexo)
    except Exception as ex:
        print("Error en", fichero_excel)
        print(ex)

# Crear un DataFrame único con la información de todos los años (1998 a 2019)
conductores_por_sexo = procesadorDFConductores.unir_conductores_x_sexo_anios(lista_conductores_x_sexo)
conductores_por_sexo
```

```python
conductores_por_sexo.to_csv('../data/clean/conductores_por_edad_sexo_1998_2019.csv', index=False)
```
